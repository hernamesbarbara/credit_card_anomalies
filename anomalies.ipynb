```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
import statsmodels.formula.api as sm
```

```python
def read_data():
	f_train = "./data/DataminingContest2009.Task1.Train.Inputs"
	f_target = "./data/DataminingContest2009.Task1.Train.Targets"
	
	df = pd.read_csv(f_train)
	targets = pd.read_csv(f_target, names=["anomaly"])
	
	df["anomaly"] = targets.anomaly
	bool_cols = [col for col in df.columns if np.all(df[col] < 2)]
	number_cols = ['amount', 'total', 'field3']
	
	for col in number_cols:
		df[col] = df[col].astype(float)
        
	for col in bool_cols:
		df[col] = df[col] == 1

	return df
```

```python
df = read_data()
print "observations: %d" % len(df)
print "anomalies: %d" % len(df[df.anomaly])
df.head()
```

```python
# convert `state1` to a factor variable
def to_factor(lab, fact):
    level = fact.levels.get_loc(lab)
    return level

states = pd.Series(sorted(df.state1.unique()))
states = pd.Factor(labels=states.index, levels=states.values, name="State")
df.state1 = df.state1.apply(lambda lab: to_factor(lab, fact=states))
```

```python
# convert `domain1` to a factor as well
domains = pd.Series(sorted(df.domain1.unique()))
domains = pd.Factor(labels=domains.index, levels=domains.values, name="Domain")
df.domain1 = df.domain1.apply( lambda x: to_factor(x, fact=domains) )
```

```python
# target variable we're going to predict is df.anomaly
# change that to a factor as well
classes = pd.factorize(df.anomaly)
classes = pd.Factor(labels=classes[0], levels=classes[1], name="Anomaly")
df.anomaly = classes.labels
df.head()
```

```python
# import numpy as np

def cartesian(arrays, out=None):
    """
    Generate a cartesian product of input arrays.

    Parameters
    ----------
    arrays : list of array-like
        1-D arrays to form the cartesian product of.
    out : ndarray
        Array to place the cartesian product in.

    Returns
    -------
    out : ndarray
        2-D array of shape (M, len(arrays)) containing cartesian products
        formed of input arrays.

    Examples
    --------
    >>> cartesian(([1, 2, 3], [4, 5], [6, 7]))
    array([[1, 4, 6],
           [1, 4, 7],
           [1, 5, 6],
           [1, 5, 7],
           [2, 4, 6],
           [2, 4, 7],
           [2, 5, 6],
           [2, 5, 7],
           [3, 4, 6],
           [3, 4, 7],
           [3, 5, 6],
           [3, 5, 7]])

    """

    arrays = [np.asarray(x) for x in arrays]
    dtype = arrays[0].dtype

    n = np.prod([x.size for x in arrays])
    if out is None:
        out = np.zeros([n, len(arrays)], dtype=dtype)

    m = n / arrays[0].size
    out[:,0] = np.repeat(arrays[0], m)
    if arrays[1:]:
        cartesian(arrays[1:], out=out[0:m,1:])
        for j in xrange(1, arrays[0].size):
            out[j*m:(j+1)*m,1:] = out[0:m,1:]
    return out
```

```python
def isolate_and_plot(variable):

    grouped = pd.pivot_table(combos, values=['pred'], rows=[variable, 'flag1'],
                             aggfunc=np.mean)

    colors = 'rbgyrbgy'
    for col in combos.flag1.unique():
        plt_data = grouped.ix[grouped.index.get_level_values(1)==col]
        pl.plot(plt_data.index.get_level_values(0), plt_data['pred'],
                color=colors[int(col)])

    pl.xlabel(variable)
    pl.ylabel("P(pred=1)")
    pl.legend(['0', '1'], loc='upper left', title='flag1')
    pl.title("Prob(pred=1) isolating " + variable + " and flag1")
    pl.show()

isolate_and_plot('flag5')
isolate_and_plot('field3')
```
